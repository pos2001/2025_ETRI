```python
#이 코드는 @hybrid_job 데코레이터를 사용하지 않고 직접 AwsQuantumJob.create()를 호출하여 하이브리드 작업을 생성하고 관리하는 방법을 보여줍니다. 

#하이브리드 잡 생성 및 관리
from braket.aws import AwsQuantumJob
from braket.devices import Devices

# This cell may take 5 minutes
# @hybrid_job 데코레이터를 사용하지 않고도 AwsQuantumJob.create()를 사용하여 하이브리드 잡 생성 가능

job = AwsQuantumJob.create(
    #양자 하드웨어로 SV 시뮬레이터 사용하겠다
    Devices.Amazon.SV1,
    
    source_module="algorithm_script.py",
    
    #작업이 완료될 때까지 기다리도록 지시
    wait_until_complete=True
)
```

```python
#완료된 작업의 결과를 가져옵니다.
results = job.result()
print("counts: ", results["counts"])
print("angle: ", results["angle"])
print("estimated cost: ", results["estimated cost"])
```
